# sum the album, artist and genre

```python
import pandas as pd
data=pd.read_csv('test2.csv')
```

```python
UserID=data['UserID_test']
TrackID=data['TrackID_test']
album=data['album_test']
artist=data['artist_test']
genre=data['genre_test']
```

```python
score=[]
for i in range(len(UserID)):
    rating=album[i]+artist[i]+genre[i]
    score.append(rating)
```

```python
print(score)
```

```python
track=[]
Score_new=[]
Score_temp=[]
count=0
for i in range(len(score)):
    count+=1
    track.append(str(UserID[i])+'_'+str(TrackID[i]))
    Score_temp.append(score[i])
    if count%6==0:        
        Score_temp.sort(reverse=True)
        for j in range(count-6,count):
            if score[j] in Score_temp[:3]:
                Score_new.append(1)
            else:
                Score_new.append(0)
        Score_temp=[]
```

```python
print(Score_new)
```

```python
dataframe = pd.DataFrame({'TrackID':track, 'Predictor':Score_new})
dataframe.to_csv("mf_sum2.csv",index=False)
```
